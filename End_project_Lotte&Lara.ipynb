```python
#Load data
life = (pd.read_csv('https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/Life%20Expectancy%20Data.csv')
.rename(columns=str.strip)) # making sure the column names don't have leading or trailing spaces
```

```python
#Change variable name: infant deaths => Infant Deaths
life = life.rename(columns={"infant deaths": "Infant Deaths"})
```

```python
#Making a subset of the dataset life with variables of interest
life_subset = life[["Country", "GDP", "Adult Mortality", "Infant Deaths", "HIV/AIDS"]]

#Making a subset of the dataset life_subset with countries of interest
life_subset_selected_countries = life_subset[(life_subset['Country'] == 'Germany') |
 (life_subset['Country'] == 'Netherlands') |
 (life_subset['Country'] == 'Algeria') |
 (life_subset['Country'] == 'Nigeria')]
```

```python
#Paragraph for dataset
text_paragraph_dataset = "Overview dataset"
text_dataset = "In this reasearch, we made use of a subset of the life dataset, based on our variables of interest. The dataset we used:"

#Create subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 0.8))

#Plot text on first subplot (without axes)
axs[0].text(0.0, 1.0, text_paragraph_dataset, color='black', fontsize=16, ha="left", va='center')
axs[0].axis('off')

#Plot normal text on second subplot (without axes)
axs[1].text(0.0, 0.99, text_dataset, color='black', fontsize=11, ha="left", va='center')
axs[1].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

# Plot dataset.head
print(life_subset_selected_countries)
```
